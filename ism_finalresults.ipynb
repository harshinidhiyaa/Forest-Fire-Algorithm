```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import random
import numpy as np

# Load the dataset
file_path = 'sample.csv'  # Update this path with the actual file path
df = pd.read_csv(file_path)

# Create a directed graph from the dataset
G = nx.DiGraph()

# Add initial edges to the graph
for index, row in df.iterrows():
    citing_patent = row['citing']
    cited_patent = row['cited']
    G.add_edge(citing_patent, cited_patent)

# Print basic information about the graph
print(f"Number of nodes: {G.number_of_nodes()}")
print(f"Number of edges: {G.number_of_edges()}")

def geometric_random(mean):
    return np.random.geometric(1 / mean)

def forest_fire_model(G, p, r, T):
    for t in range(1, T + 1):
        new_node = max(G.nodes) + 1
        G.add_node(new_node)
        
        # Select an ambassador node uniformly at random
        ambassador_node = random.choice(list(G.nodes))
        G.add_edge(new_node, ambassador_node)
        
        # Generate geometric random numbers
        x = geometric_random((1 - p) ** -1)
        y = geometric_random((1 - r * p) ** -1)
        
        # Select out-links and in-links to burn
        out_links = list(G.out_edges(ambassador_node))
        in_links = list(G.in_edges(ambassador_node))
        
        selected_links = random.sample(out_links, min(x, len(out_links))) + random.sample(in_links, min(y, len(in_links)))
        
        for edge in selected_links:
            target_node = edge[1] if edge in out_links else edge[0]
            if not G.has_edge(new_node, target_node):
                G.add_edge(new_node, target_node)
        
        # Recursive step
        for edge in selected_links:
            target_node = edge[1] if edge in out_links else edge[0]
            forest_fire_recursive(G, new_node, target_node, p, r)

def forest_fire_recursive(G, current_node, target_node, p, r):
    if target_node in G.neighbors(current_node):
        return

    G.add_edge(current_node, target_node)
    
    # Generate geometric random numbers
    x = geometric_random((1 - p) ** -1)
    y = geometric_random((1 - r * p) ** -1)
    
    out_links = list(G.out_edges(target_node))
    in_links = list(G.in_edges(target_node))
    
    selected_links = random.sample(out_links, min(x, len(out_links))) + random.sample(in_links, min(y, len(in_links)))
    
    for edge in selected_links:
        next_target_node = edge[1] if edge in out_links else edge[0]
        if not G.has_edge(current_node, next_target_node):
            G.add_edge(current_node, next_target_node)
        forest_fire_recursive(G, current_node, next_target_node, p, r)

# Parameters for the Forest Fire model
p = 0.35  # Forward burning probability
r = 0.32  # Backward burning ratio
T = 100   # Number of time steps

# Apply the Forest Fire model to the graph
forest_fire_model(G, p, r, T)

# Analyze the resulting graph
print(f"Number of nodes after densification: {G.number_of_nodes()}")
print(f"Number of edges after densification: {G.number_of_edges()}")
degree_distribution = [G.degree(n) for n in G.nodes]

# Plot degree distribution
plt.figure(figsize=(10, 6))
plt.hist(degree_distribution, bins=range(1, max(degree_distribution) + 1), edgecolor='black')
plt.yscale('log')
plt.xscale('log')
plt.title('Degree Distribution (Log-Log Scale)')
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.show()

# Calculate and print the average path length of the largest weakly connected component
largest_wcc = max(nx.weakly_connected_components(G), key=len)
subgraph_wcc = G.subgraph(largest_wcc).to_undirected()
if nx.is_connected(subgraph_wcc):
    avg_path_length = nx.average_shortest_path_length(subgraph_wcc)
    print(f'Average Path Length: {avg_path_length}')
else:
    print('The largest weakly connected component is not connected, so average path length cannot be computed.')

# Print the diameter of the largest strongly connected component
largest_scc = max(nx.strongly_connected_components(G), key=len)
subgraph_scc = G.subgraph(largest_scc)
if nx.is_strongly_connected(subgraph_scc):
    diameter = nx.diameter(subgraph_scc.to_undirected())
    print(f'Diameter of the largest strongly connected component: {diameter}')
else:
    print('The largest strongly connected component is not strongly connected, so diameter cannot be computed.')

# Plot the graph
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=False, node_size=10, node_color="blue", edge_color="gray", alpha=0.7)
plt.title("Densified Patent Citation Network")
plt.show()
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import random
import numpy as np

# Load the dataset
file_path = 'subset_dataset3.csv'  # Update this path with the actual file path
df = pd.read_csv(file_path)

# Ensure the CSV has the necessary columns
required_columns = ['Patent', 'Citation', 'Cit_Date']
assert all(col in df.columns for col in required_columns), f"CSV must contain columns: {required_columns}"

# Create a directed graph from the dataset
G = nx.DiGraph()

# Add initial edges to the graph
for index, row in df.iterrows():
    citing_patent = str(row['Patent'])  # Ensure patent IDs are treated as strings
    cited_patent = str(row['Citation'])
    G.add_edge(citing_patent, cited_patent)

# Print basic information about the graph
print(f"Number of nodes: {G.number_of_nodes()}")
print(f"Number of edges: {G.number_of_edges()}")

def geometric_random(mean):
    return np.random.geometric(1 / mean)

def forest_fire_model(G, p, r, T):
    for t in range(1, T + 1):
        new_node = str(max(map(int, G.nodes)) + 1)  # Generate new node as string
        G.add_node(new_node)
        
        # Select an ambassador node uniformly at random
        ambassador_node = random.choice(list(G.nodes))
        G.add_edge(new_node, ambassador_node)
        
        # Generate geometric random numbers
        x = geometric_random((1 - p) ** -1)
        y = geometric_random((1 - r * p) ** -1)
        
        # Select out-links and in-links to burn
        out_links = list(G.out_edges(ambassador_node))
        in_links = list(G.in_edges(ambassador_node))
        
        selected_links = random.sample(out_links, min(x, len(out_links))) + random.sample(in_links, min(y, len(in_links)))
        
        for edge in selected_links:
            target_node = edge[1] if edge in out_links else edge[0]
            if not G.has_edge(new_node, target_node):
                G.add_edge(new_node, target_node)
        
        # Recursive step
        for edge in selected_links:
            target_node = edge[1] if edge in out_links else edge[0]
            forest_fire_recursive(G, new_node, target_node, p, r)

def forest_fire_recursive(G, current_node, target_node, p, r):
    if G.has_edge(current_node, target_node):
        return

    G.add_edge(current_node, target_node)
    
    # Generate geometric random numbers
    x = geometric_random((1 - p) ** -1)
    y = geometric_random((1 - r * p) ** -1)
    
    out_links = list(G.out_edges(target_node))
    in_links = list(G.in_edges(target_node))
    
    selected_links = random.sample(out_links, min(x, len(out_links))) + random.sample(in_links, min(y, len(in_links)))
    
    for edge in selected_links:
        next_target_node = edge[1] if edge in out_links else edge[0]
        if not G.has_edge(current_node, next_target_node):
            G.add_edge(current_node, next_target_node)
        forest_fire_recursive(G, current_node, next_target_node, p, r)

# Parameters for the Forest Fire model
p = 0.35  # Forward burning probability
r = 0.32  # Backward burning ratio
T = 100   # Number of time steps

# Apply the Forest Fire model to the graph
forest_fire_model(G, p, r, T)

# Analyze the resulting graph
print(f"Number of nodes after densification: {G.number_of_nodes()}")
print(f"Number of edges after densification: {G.number_of_edges()}")

# Degree distributions
in_degree_dist = [G.in_degree(n) for n in G.nodes()]
out_degree_dist = [G.out_degree(n) for n in G.nodes()]

# Calculate the frequency of each in-degree and out-degree
in_degree_freq = {degree: in_degree_dist.count(degree) for degree in set(in_degree_dist)}
out_degree_freq = {degree: out_degree_dist.count(degree) for degree in set(out_degree_dist)}

# Convert to lists for plotting
in_degree_x = list(in_degree_freq.keys())
in_degree_y = list(in_degree_freq.values())
out_degree_x = list(out_degree_freq.keys())
out_degree_y = list(out_degree_freq.values())

# Plot degree distributions as scatter plots
plt.figure(figsize=(12, 6))

# In-degree distribution
plt.subplot(1, 2, 1)
plt.scatter(in_degree_x, in_degree_y, edgecolor='black')
plt.yscale('log')
plt.xscale('log')
plt.title('In-Degree Distribution (Log-Log Scale)')
plt.xlabel('Degree')
plt.ylabel('Frequency')

# Out-degree distribution
plt.subplot(1, 2, 2)
plt.scatter(out_degree_x, out_degree_y, edgecolor='black')
plt.yscale('log')
plt.xscale('log')
plt.title('Out-Degree Distribution (Log-Log Scale)')
plt.xlabel('Degree')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Calculate Effective Diameter Over Time
df['year'] = pd.to_datetime(df['Cit_Date']).dt.year

def calculate_effective_diameter_over_time(graph, df):
    years = sorted(df['year'].unique())
    effective_diameters = []

    for year in years:
        subgraph_nodes = [node for node, data in graph.nodes(data=True) if int(node) in df[df['year'] <= year]['Patent'].astype(int).tolist()]
        subgraph = graph.subgraph(subgraph_nodes)
        if nx.is_connected(subgraph.to_undirected()):
            effective_diameters.append(nx.diameter(subgraph.to_undirected()))
        else:
            effective_diameters.append(None)  # Handle disconnected subgraphs

    return years, effective_diameters

years, diameters = calculate_effective_diameter_over_time(G, df)

# Plot Time vs Effective Diameter
plt.figure(figsize=(10, 6))
plt.plot(years, diameters, marker='o', linestyle='-', color='b')
plt.title('Time vs Effective Diameter')
plt.xlabel('Time (Year)')
plt.ylabel('Effective Diameter')
plt.grid(True)
plt.show()
```
